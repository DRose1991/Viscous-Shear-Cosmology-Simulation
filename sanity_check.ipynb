```python
# @title VSC Validation: Residuals & Pre-Registration
import numpy as np
import matplotlib.pyplot as plt

# --- CONFIGURATION ---
H0_Planck = 67.4
H0_Shoes = 73.0
Omega_m = 0.315
Omega_L_static = 0.685
nu = 0.08  # The VSC Viscosity Index

# --- 1. RESIDUAL ANALYSIS (ΔH/H) ---
z = np.linspace(0, 5, 200)

# Calculate H(z) for both models
E_lcdm = np.sqrt(Omega_m*(1+z)**3 + Omega_L_static)
E_vsc  = np.sqrt(Omega_m*(1+z)**3 + Omega_L_static*(1+z)**nu)

# Fractional Difference (Residual)
residual = (E_vsc - E_lcdm) / E_lcdm

# Plot
plt.style.use('dark_background')
plt.figure(figsize=(8, 4))
plt.plot(z, residual, 'y-', linewidth=2)
plt.axhline(0, color='red', linestyle='--', alpha=0.5)
plt.title("Residuals: VSC Deviation from ΛCDM", color='white')
plt.ylabel("ΔH / H_lcdm", color='white')
plt.xlabel("Redshift (z)", color='white')
plt.grid(True, alpha=0.3)
plt.show()

# --- 2. PRE-REGISTERED PREDICTIONS ---
# Prediction: Effective Dark Energy Density relative to Today
def get_eff_de(z_target):
    return Omega_L_static * (1+z_target)**nu

print(f"--- PRE-REGISTERED PREDICTIONS (nu={nu}) ---")
print(f"z=2.0:  {get_eff_de(2.0):.4f}")
print(f"z=5.0:  {get_eff_de(5.0):.4f}")
print(f"z=10.0: {get_eff_de(10.0):.4f}")
```
