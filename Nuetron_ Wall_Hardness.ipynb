```python
# @title Ink Drive: Metric Hardness Test (Stress-Strain)
import numpy as np
import matplotlib.pyplot as plt

# --- CONFIGURATION ---
# We simulate the "Feel" of the vacuum as we ramp up the drive.
shear_rate = np.logspace(-2, 1, 500) # From 0.01 to 10 (normalized to 1/tau)
tau_vac = 1.0  # Critical time
n = 20         # Dilatancy Index (The "Ink Factor")
zeta_0 = 1e-3  # Base viscosity

# --- PHYSICS ---
# Effective Viscosity: zeta = zeta_0 * (1 + (rate * tau)^n)
viscosity = zeta_0 * (1 + (shear_rate * tau_vac)**n)

# Shear Stress (The "Resistance" you feel): sigma = viscosity * rate
stress = viscosity * shear_rate

# --- VISUALIZATION ---
plt.style.use('dark_background')
plt.figure(figsize=(10, 6))

plt.loglog(shear_rate, stress, 'm-', linewidth=3)
plt.axvline(1.0, color='cyan', linestyle='--', label='Ink Limit (Ik=1)')

# Annotations
plt.title("Vacuum Stress-Strain Curve: How Hard is the Wall?", fontsize=16, color='white')
plt.xlabel("Drive Frequency (Impact Speed)", fontsize=12, color='white')
plt.ylabel("Metric Resistance (Shear Stress)", fontsize=12, color='white')

plt.text(0.05, 1e-4, "LIQUID PHASE\n(Soft Vacuum)\nShip Sinks", color='cyan', fontsize=12)
plt.text(2.0, 1e15, "SOLID PHASE\n(Neutron Wall)\nShip Skips", color='magenta', fontsize=12)

plt.grid(True, which="both", alpha=0.3)
plt.legend()
plt.show()
```
