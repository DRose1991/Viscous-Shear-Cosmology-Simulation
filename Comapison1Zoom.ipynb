```python
# @title VSC vs CDM: The "Microscope" View (Zoom In)
# ==============================================================================
# Focuses only on late-time expansion to visualize the divergence.
# ==============================================================================

import matplotlib.pyplot as plt
import numpy as np

# Create the figure
plt.style.use('dark_background')
plt.figure(figsize=(10, 6))

# Zoom in on the range z = 0 to z = 0.5 (The recent universe)
z_zoom = np.linspace(0, 0.5, 50)

# Recalculate for just this slice
# LCDM (Static)
H_lcdm_zoom = 67.4 * np.sqrt(0.315 * (1+z_zoom)**3 + 0.685)

# VSC (Dynamic) - using the nu=0.08 we found worked
H_vsc_zoom = 67.4 * np.sqrt(0.315 * (1+z_zoom)**3 + 0.685 * (1+z_zoom)**0.08)

# Plot the difference
plt.plot(z_zoom, H_lcdm_zoom, 'r--', linewidth=2, label='CDM (Goes to 67.4)')
plt.plot(z_zoom, H_vsc_zoom, 'c-', linewidth=4, alpha=0.6, label='VSC (Goes to 73.0)')

# Fill the gap between them to highlight the "Viscous Boost"
plt.fill_between(z_zoom, H_lcdm_zoom, H_vsc_zoom, color='cyan', alpha=0.2, label='The Viscous Drift')

# Add the Targets
plt.scatter([0], [73.0], color='lime', s=200, zorder=5, label='Target (SH0ES)')
plt.scatter([0], [67.4], color='red', marker='x', s=200, zorder=5, label='Anchor (Planck)')

# Formatting
plt.title("The Divergence: Last 5 Billion Years", fontsize=16, color='white')
plt.xlabel("Redshift (z)", fontsize=12, color='white')
plt.ylabel("Expansion Rate H(z)", fontsize=12, color='white')
plt.xlim(0.5, 0) # Flip axis so Time goes Left -> Right (Past -> Future)
plt.grid(True, alpha=0.3)
plt.legend()

plt.show()
```
